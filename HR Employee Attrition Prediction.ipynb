```python
# Algorithms used for classification:
1. LogisticRegression
2. DecisionTreeClassifier
3. RandomForestClassifier
4. Naive_bayes
5. KNeighborsClassifier
```

```python
# Splits the dataset into features (x) by dropping the 'fraud' column and target variable (y) as 'fraud'.
x = df.drop('Attrition',axis=1)
y = df['Attrition']
```

```python
print(x)
print("---"*25)
print(y)
```

```python
y = np.array(y)
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Standardizes the features (x) to have a mean of 0 and a standard deviation of 1.
ss = StandardScaler()
ss.fit(x)
```

```python
ss.transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.20,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
print('Confusion Matrix:\n',confusion_matrix(y_test,y_pred))
print('\nClassification Report:\n',classification_report(y_test,y_pred))
print('\nAccuracy Score:\n',accuracy_score(y_test,y_pred))
```

```python
#####
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier = DecisionTreeClassifier()
classifier.fit(x_train,y_train)
```

```python
y_pred1 = classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
print('Confusion Matrix:\n',confusion_matrix(y_test,y_pred1))
print('\nClassification Report:\n',classification_report(y_test,y_pred1))
print('\nAccuracy Score:\n',accuracy_score(y_test,y_pred1))
```

```python
#####
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier = RandomForestClassifier(n_estimators = 20,random_state=0)
classifier.fit(x_train,y_train)
```

```python
y_pred2 = classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
print('Confusion Matrix:\n',confusion_matrix(y_test,y_pred2))
print('\nClassification Report:\n',classification_report(y_test,y_pred2))
print('\nAccuracy Score:\n',accuracy_score(y_test,y_pred2))
```
